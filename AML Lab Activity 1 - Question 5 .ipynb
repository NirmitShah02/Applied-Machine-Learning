```python
data['heartRate'] = data['heartRate'].astype('category')
```

```python
data['heartRate'] = pd.cut(data.heartRate, [44, 75, 143], labels = ["High Risk", "Low Risk"])
```

```python
data
```

```python
data.isnull().sum() #checking whether all missing values are dropped or no
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum() #checking whether all missing values are dropped or no
```

```python
data
```

```python
# independent variable
in_var = ['age', 'BMI']
X=data[in_var]
# dependent variable
y=data.heartRate
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(X)
```

```python
X = scaler.transform(X)
```

```python
X
```

```python
y.value_counts()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=7) 
#splitted dataset into train and test
```

```python
X_train.shape #no. of rows and columns in training data in X-axis
```

```python
X_test.shape #no. of rows and columns in testing data in X-axis
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
lr = LogisticRegression() #instantiated logistic regression model
```

```python
#fitting model with training data
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
y_pred #predicted values
```

```python
y_test #actual values
```

```python
con_mat = metrics.confusion_matrix(y_test, y_pred)  #confusion matrix
print(con_mat)
```

```python
print('Accuracy: ',metrics.accuracy_score(y_test, y_pred)*100,'%')
```

```python
print(classification_report(y_test, y_pred))
```
