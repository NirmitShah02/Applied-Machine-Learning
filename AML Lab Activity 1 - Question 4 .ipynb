#### Implement  a  Decision  Tree  model  in  Machine  Learningand  fit  the  model  towards  the regression to predict housing price. Use “melb_data.csv” dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error

from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("melb_data.csv")

data
```

```python
data= data.dropna()
data = pd.concat([data, pd.get_dummies(data["Type"]), pd.get_dummies(data["Method"]), pd.get_dummies(data["Regionname"])], axis=1)
data = data.drop(["Suburb", "Address", "SellerG", "CouncilArea", "Type", "Method", "Regionname"], 1)
data['Date'] = [pd.Timestamp(x).timestamp() for x in data["Date"]]
data = data.dropna()
data.head()
```

```python
X = data.drop("Price", 1)
```

```python
y = data['Price']
```

```python
y
```

```python
X = data.drop(["Price","Date"], 1)
X
```

```python
u = np.mean(X)
std = np.std(X, axis = 0)
X = (X-u)/std
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
rescaledX = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state=0)
```

```python
model = DecisionTreeRegressor(random_state=10)

model.fit(X_train,y_train)
```

```python
y_pred_dt = model.predict(X_test)
```

```python
print ("Prediction of testing data by decision tree for regression:", y_pred_dt)
```

```python
from sklearn.metrics import accuracy_score
accuracy_dt = model.score(X_test,y_test)
print("Accuracy of Decision Tree for Regression Model in %:",accuracy_dt*100)
```
